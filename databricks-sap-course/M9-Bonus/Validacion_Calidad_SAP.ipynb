```python
# Validación de calidad post-pipeline — falla el Job si hay demasiadas filas en cuarentena

umbral_maximo = 100  # ajusta según tu tolerancia de negocio

resultado = spark.sql("""
    SELECT COUNT(*) AS total
    FROM laboratory_sap_dev.lakeflow_demo.silver_trips_quarantine
""").collect()[0]['total']

print(f"Filas en cuarentena: {resultado}")

if resultado > umbral_maximo:
    raise Exception(
        f"⚠️ ALERTA CALIDAD DE DATOS: {resultado} filas en cuarentena "
        f"(umbral: {umbral_maximo}). Revisar origen antes de continuar."
    )
else:
    print(f"✅ Calidad OK: {resultado} filas en cuarentena, dentro del umbral.")
```
